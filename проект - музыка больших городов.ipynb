–ü—Ä–æ–≤–µ—Ä—å—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –î–ª—è —ç—Ç–æ–≥–æ –µ—â—ë —Ä–∞–∑ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤:

````python
# –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ - –ø–µ—Ä–µ—á–µ–Ω—å –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤

print(df.columns)
````

–ù–µ –≤—Å–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤–ª–∏—è—é—Ç –Ω–∞ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ. –¢–∞–∫ –≤ `track` –∏ `artist` –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–µ –≤–∞–∂–Ω—ã –¥–ª—è –≤–∞—à–µ–π —Ä–∞–±–æ—Ç—ã. –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –∏—Ö —è–≤–Ω—ã–º–∏ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è–º–∏.

–ù–æ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ `genre` –º–æ–≥—É—Ç –ø–æ–º–µ—à–∞—Ç—å —Å—Ä–∞–≤–Ω–µ–Ω–∏—é –º—É–∑—ã–∫–∞–ª—å–Ω—ã—Ö –≤–∫—É—Å–æ–≤ –≤ –ú–æ—Å–∫–≤–µ –∏ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ. –ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –±—ã–ª–æ –±—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –ø—Ä–∏—á–∏–Ω—É –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –¥–∞–Ω–Ω—ã–µ. –¢–∞–∫–æ–π –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –Ω–µ—Ç –≤ —É—á–µ–±–Ω–æ–º –ø—Ä–æ–µ–∫—Ç–µ. –ü—Ä–∏–¥—ë—Ç—Å—è:
* –∑–∞–ø–æ–ª–Ω–∏—Ç—å –∏ —ç—Ç–∏ –ø—Ä–æ–ø—É—Å–∫–∏ —è–≤–Ω—ã–º–∏ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è–º–∏,
* –æ—Ü–µ–Ω–∏—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ –æ–Ω–∏ –ø–æ–≤—Ä–µ–¥—è—Ç —Ä–∞—Å—á—ë—Ç–∞–º. 

–ó–∞–º–µ–Ω–∏—Ç–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö `track`, `artist` –∏ `genre` –Ω–∞ —Å—Ç—Ä–æ–∫—É `'unknown'`. –î–ª—è —ç—Ç–æ–≥–æ —Å–æ–∑–¥–∞–π—Ç–µ —Å–ø–∏—Å–æ–∫ `columns_to_replace`, –ø–µ—Ä–µ–±–µ—Ä–∏—Ç–µ –µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç—ã —Ü–∏–∫–ª–æ–º `for` –∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –∑–∞–º–µ–Ω—É –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:

````python
# –ø–µ—Ä–µ–±–æ—Ä –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ —Ü–∏–∫–ª–µ –∏ –∑–∞–º–µ–Ω–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ 'unknown'

columns_to_replace = ['track', 'artist', 'genre']

for column in columns_to_replace:
    df[column] = df[column].fillna('unknown')
````

–£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤ —Ç–∞–±–ª–∏—Ü–µ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å –ø—Ä–æ–ø—É—Å–∫–æ–≤. –î–ª—è —ç—Ç–æ–≥–æ –µ—â—ë —Ä–∞–∑ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

````python
# –ø–æ–¥—Å—á—ë—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤

print(df.isna().sum())
````

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ —Ç–∞–±–ª–∏—Ü–µ –æ–¥–Ω–æ–π –∫–æ–º–∞–Ω–¥–æ–π:

````python
# —É–¥–∞–ª–µ–Ω–∏–µ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ (—Å —É–¥–∞–ª–µ–Ω–∏–µ–º —Å—Ç–∞—Ä—ã—Ö –∏–Ω–¥–µ–∫—Å–æ–≤ –∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ–º –Ω–æ–≤—ã—Ö)

df = df.drop_duplicates().reset_index(drop=True)
````

–ï—â—ë —Ä–∞–∑ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ —Ç–∞–±–ª–∏—Ü–µ ‚Äî —É–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ—Ç –Ω–∏—Ö –∏–∑–±–∞–≤–∏–ª–∏—Å—å:

````python
# –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

print(df.duplicated().sum())
````

–¢–µ–ø–µ—Ä—å –∏–∑–±–∞–≤—å—Ç–µ—Å—å –æ—Ç –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –∫–æ–ª–æ–Ω–∫–µ `genre`. –ù–∞–ø—Ä–∏–º–µ—Ä, –Ω–∞–∑–≤–∞–Ω–∏–µ –æ–¥–Ω–æ–≥–æ –∏ —Ç–æ–≥–æ –∂–µ –∂–∞–Ω—Ä–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –∑–∞–ø–∏—Å–∞–Ω–æ –Ω–µ–º–Ω–æ–≥–æ –ø–æ-—Ä–∞–∑–Ω–æ–º—É. –¢–∞–∫–∏–µ –æ—à–∏–±–∫–∏ —Ç–æ–∂–µ –ø–æ–≤–ª–∏—è—é—Ç –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è.

–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π –∂–∞–Ω—Ä–æ–≤, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤ –∞–ª—Ñ–∞–≤–∏—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ. –î–ª—è —ç—Ç–æ–≥–æ:
* –∏–∑–≤–ª–µ–∫–∏—Ç–µ –Ω—É–∂–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, 
* –ø—Ä–∏–º–µ–Ω–∏—Ç–µ –∫ –Ω–µ–º—É –º–µ—Ç–æ–¥ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏,
* –¥–ª—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤—ã–∑–æ–≤–∏—Ç–µ –º–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –≤–µ—Ä–Ω—ë—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ —Å—Ç–æ–ª–±—Ü–∞.

````python
# –ü—Ä–æ—Å–º–æ—Ç—Ä —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π –∂–∞–Ω—Ä–æ–≤

print(df['genre'].sort_values().unique())
````

```python
sorted(df['genre'].unique())

```

–¢–æ—á–Ω–æ. –ü–æ—á–µ–º—É —Ç–æ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –∑–∞–¥–∞–Ω–∏—è –Ω–µ –æ–±—Ä–∞—Ç–∏–ª–∞ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç–æ, —á—Ç–æ –Ω—É–∂–Ω–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞. –ò—Å–ø—Ä–∞–≤–∏–ª–∞.
</div>

–ü—Ä–æ—Å–º–æ—Ç—Ä–∏—Ç–µ —Å–ø–∏—Å–æ–∫ –∏ –Ω–∞–π–¥–∏—Ç–µ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –Ω–∞–∑–≤–∞–Ω–∏—è `hiphop`. –≠—Ç–æ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è —Å –æ—à–∏–±–∫–∞–º–∏ –∏–ª–∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Ç–æ–≥–æ –∂–µ –∂–∞–Ω—Ä–∞.

–í—ã —É–≤–∏–¥–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã:
* *hip*,
* *hop*,
* *hip-hop*.

–ß—Ç–æ–±—ã –æ—á–∏—Å—Ç–∏—Ç—å –æ—Ç –Ω–∏—Ö —Ç–∞–±–ª–∏—Ü—É, –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `replace_wrong_genres()` —Å –¥–≤—É–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: 
* `wrong_genres` ‚Äî —Å–ø–∏—Å–æ–∫ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤,
* `correct_genre` ‚Äî —Å—Ç—Ä–æ–∫–∞ —Å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º.

–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –∏—Å–ø—Ä–∞–≤–∏—Ç—å –∫–æ–ª–æ–Ω–∫—É `genre` –≤ —Ç–∞–±–ª–∏—Ü–µ `df`: –∑–∞–º–µ–Ω–∏—Ç—å –∫–∞–∂–¥–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å–ø–∏—Å–∫–∞ `wrong_genres` –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ `correct_genre`.

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–º–µ–Ω—ã –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

def replace_wrong_genres(wrong_genres, correct_genre):
    for wrong_genre in wrong_genres:
        df['genre'] = df['genre'].replace(wrong_genre, correct_genre)
````

–í—ã–∑–æ–≤–∏—Ç–µ `replace_wrong_genres()` –∏ –ø–µ—Ä–µ–¥–∞–π—Ç–µ –µ–π —Ç–∞–∫–∏–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã, —á—Ç–æ–±—ã –æ–Ω–∞ —É—Å—Ç—Ä–∞–Ω–∏–ª–∞ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã: –≤–º–µ—Å—Ç–æ `hip`, `hop` –∏ `hip-hop` –≤ —Ç–∞–±–ª–∏—Ü–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ `hiphop`:

````python
# –£—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

duplicates = ['hip', 'hop', 'hip-hop']
name = 'hiphop'
replace_wrong_genres(duplicates, name)
````

–í–æ–æ–±—â–µ, —ç—Ç–∏ –º–∞–Ω–∏–ø—É–ª—è—Ü–∏–∏ –º–æ–∂–Ω–æ –ø—Ä–æ–∏–∑–≤–µ—Å—Ç–∏ –∏ –±–µ–∑ —Ñ—É–Ω–∫—Ü–∏–∏. –¢—ã –≤–µ—Ä–Ω–æ –¥–µ–ª–∞–µ—à—å, —á—Ç–æ —Å–ª–µ–¥—É–µ—à—å –¢–ó, –Ω–æ —è —Ö–æ—á—É –ø–æ–∫–∞–∑–∞—Ç—å –±–æ–ª–µ–µ –ª–∞–∫–æ–Ω–∏—á–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–º –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è —Å–ø–∏—Å–æ–∫ –∏ –≤—Å–µ —Ç–æ—Ç –∂–µ –º–µ—Ç–æ–¥ **replace**.   –í—ã–≥–ª—è–¥–µ—Ç—å —ç—Ç–æ –±—É–¥–µ—Ç —Ç–∞–∫:

–ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –∑–∞–º–µ–Ω–∏–ª–∏ –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è:

*   hip
*   hop
*   hip-hop

–í—ã–≤–µ–¥–∏—Ç–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ `genre`:

````python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã

print(df['genre'].sort_values().unique())
````

- –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è,

* –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏,
* –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞.

  * –≤ –∫–æ–ª–æ–Ω–∫–µ `day` —Ä–∞–≤–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É `day`,
  * –≤ –∫–æ–ª–æ–Ω–∫–µ `city` —Ä–∞–≤–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É `city`.

–î–ª—è —ç—Ç–æ–≥–æ –ø—Ä–∏–º–µ–Ω–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—É—é —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—é —Å –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–µ–π.

–ó–∞—Ç–µ–º –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ `user_id` –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è —Ç–∞–±–ª–∏—Ü—ã. –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é. –í–µ—Ä–Ω–∏—Ç–µ —ç—Ç—É –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –∏–∑ —Ñ—É–Ω–∫—Ü–∏–∏.

````python
# <—Å–æ–∑–¥–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ number_tracks()>
# –û–±—ä—è–≤–ª—è–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è —Å –¥–≤—É–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: day, city.
# –í –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π track_list —Å–æ—Ö—Ä–∞–Ω—è—é—Ç—Å—è —Ç–µ —Å—Ç—Ä–æ–∫–∏ —Ç–∞–±–ª–∏—Ü—ã df, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö 
# –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ 'day' —Ä–∞–≤–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É day –∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ
# –≤ —Å—Ç–æ–ª–±—Ü–µ 'city' —Ä–∞–≤–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É city (–∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—É—é —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—é
# —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏).
# –í –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π track_list_count —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è —á–∏—Å–ª–æ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ 'user_id',
# —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–æ–µ –º–µ—Ç–æ–¥–æ–º count() –¥–ª—è —Ç–∞–±–ª–∏—Ü—ã track_list.
# –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ - –∑–Ω–∞—á–µ–Ω–∏–µ track_list_count.

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –ø—Ä–æ—Å–ª—É—à–∏–≤–∞–Ω–∏–π –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ –≥–æ—Ä–æ–¥–∞ –∏ –¥–Ω—è.
# –° –ø–æ–º–æ—â—å—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ–π —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ —Å –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–µ–π –æ–Ω–∞
# —Å–Ω–∞—á–∞–ª–∞ –ø–æ–ª—É—á–∏—Ç –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã —Å—Ç—Ä–æ–∫–∏ —Å –Ω—É–∂–Ω—ã–º –¥–Ω–µ–º,
# –∑–∞—Ç–µ–º –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ—Ç —Å—Ç—Ä–æ–∫–∏ —Å –Ω—É–∂–Ω—ã–º –≥–æ—Ä–æ–¥–æ–º,
# –º–µ—Ç–æ–¥–æ–º count() –¥–ª—è —Ç–∞–±–ª–∏—Ü—ã track_list.
# –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ - –∑–Ω–∞—á–µ–Ω–∏–µ track_list_count.

def number_tracks(day, city):
    track_list = df[df['day'] == day]
    track_list = track_list[track_list['city'] == city]
    track_list_count = track_list['user_id'].count()
    return track_list_count
````

````python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–ª—É—à–∏–≤–∞–Ω–∏–π –≤ –ú–æ—Å–∫–≤–µ –ø–æ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞–º

number_tracks('Monday', 'Moscow')
````

–°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ç–∞–±–ª–∏—Ü—ã —Å –¥–∞–Ω–Ω—ã–º–∏ –≤ –¥–≤–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ:

–°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `genre_weekday()` —Å —á–µ—Ç—ã—Ä—å–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏:
* —Ç–∞–±–ª–∏—Ü–∞ (–¥–∞—Ç–∞—Ñ—Ä–µ–π–º) —Å –¥–∞–Ω–Ω—ã–º–∏,
* –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏,
* –Ω–∞—á–∞–ª—å–Ω–∞—è –≤—Ä–µ–º–µ–Ω–Ω–∞—è –º–µ—Ç–∫–∞ –≤ —Ñ–æ—Ä–º–∞—Ç–µ 'hh:mm', 
* –ø–æ—Å–ª–µ–¥–Ω—è—è –≤—Ä–µ–º–µ–Ω–Ω–∞—è –º–µ—Ç–∫–∞ –≤ —Ñ–æ—Ä–º–∞—Ç–µ 'hh:mm'.

–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–æ–ø-10 –∂–∞–Ω—Ä–æ–≤ —Ç–µ—Ö —Ç—Ä–µ–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ—Å–ª—É—à–∏–≤–∞–ª–∏ –≤ —É–∫–∞–∑–∞–Ω–Ω—ã–π –¥–µ–Ω—å, –≤ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ –º–µ–∂–¥—É –¥–≤—É–º—è –æ—Ç–º–µ—Ç–∫–∞–º–∏ –≤—Ä–µ–º–µ–Ω–∏.

````python
# –û–±—ä—è–≤–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ genre_weekday() —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ table, day, time1, time2,
# –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–∞—Ö –≤ —É–∫–∞–∑–∞–Ω–Ω—ã–π –¥–µ–Ω—å –≤
# –∑–∞–¥–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è:
# 1) –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é genre_df —Å–æ—Ö—Ä–∞–Ω—è—é—Ç—Å—è —Ç–µ —Å—Ç—Ä–æ–∫–∏ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ table, –¥–ª—è
#    –∫–æ—Ç–æ—Ä—ã—Ö –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ:
#    - –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ day —Ä–∞–≤–Ω–æ –∑–Ω–∞—á–µ–Ω–∏—é –∞—Ä–≥—É–º–µ–Ω—Ç–∞ day
#    - –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ time –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ä–≥—É–º–µ–Ω—Ç–∞ time1
#    - –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ time –º–µ–Ω—å—à–µ –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ä–≥—É–º–µ–Ω—Ç–∞ time2
#    –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—É—é —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—é —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏.
# 2) —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞—Ç—å –¥–∞—Ç–∞—Ñ—Ä–µ–π–º genre_df –ø–æ —Å—Ç–æ–ª–±—Ü—É genre, –≤–∑—è—Ç—å –æ–¥–∏–Ω –∏–∑ –µ–≥–æ
#    —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å –º–µ—Ç–æ–¥–æ–º count() –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑
#    –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—â–∏—Ö –∂–∞–Ω—Ä–æ–≤, –ø–æ–ª—É—á–∏–≤—à–∏–π—Å—è Series –∑–∞–ø–∏—Å–∞—Ç—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
#    genre_df_count
# 3) –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å genre_df_count –ø–æ —É–±—ã–≤–∞–Ω–∏—é –≤—Å—Ç—Ä–µ—á–∞–µ–º–æ—Å—Ç–∏ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å
#    –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é genre_df_sorted
# 4) –≤–µ—Ä–Ω—É—Ç—å Series –∏–∑ 10 –ø–µ—Ä–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π genre_df_sorted, —ç—Ç–æ –±—É–¥—É—Ç —Ç–æ–ø-10
#    –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤ (–≤ —É–∫–∞–∑–∞–Ω–Ω—ã–π –¥–µ–Ω—å, –≤ –∑–∞–¥–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è)

def genre_weekday(table, day, time1, time2):
    genre_df = table[table['day'] == day]
    genre_df = genre_df[genre_df['time'] > time1]
    genre_df = genre_df[genre_df['time'] < time2]
    genre_df_count = genre_df.groupby('genre')['user_id'].count()
    genre_df_sorted = genre_df_count.sort_values(ascending=False)
    
    return genre_df_sorted.head(10)
````

C—Ä–∞–≤–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏ `genre_weekday()` –¥–ª—è –ú–æ—Å–∫–≤—ã –∏ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–∞ –≤ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ —É—Ç—Ä–æ–º (—Å 7:00 –¥–æ 11:00) –∏ –≤ –ø—è—Ç–Ω–∏—Ü—É –≤–µ—á–µ—Ä–æ–º (—Å 17:00 –¥–æ 23:00):

````python
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —É—Ç—Ä–∞ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞ –≤ –ú–æ—Å–∫–≤–µ (–≤–º–µ—Å—Ç–æ df ‚Äî —Ç–∞–±–ª–∏—Ü–∞ moscow_general)
# –æ–±—ä–µ–∫—Ç—ã, —Ö—Ä–∞–Ω—è—â–∏–µ –≤—Ä–µ–º—è, —è–≤–ª—è—é—Ç—Å—è —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞—é—Ç—Å—è –∫–∞–∫ —Å—Ç—Ä–æ–∫–∏
# –ø—Ä–∏–º–µ—Ä –≤—ã–∑–æ–≤–∞: genre_weekday(moscow_general, 'Monday', '07:00', '11:00')

genre_weekday(moscow_general, 'Monday', '07:00', '11:00')
````

````python
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —É—Ç—Ä–∞ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞ –≤ –ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ (–≤–º–µ—Å—Ç–æ df ‚Äî —Ç–∞–±–ª–∏—Ü–∞ spb_general)

genre_weekday(spb_general, 'Monday', '07:00', '11:00')
````

````python
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤–µ—á–µ—Ä–∞ –ø—è—Ç–Ω–∏—Ü—ã –≤ –ú–æ—Å–∫–≤–µ

genre_weekday(moscow_general, 'Friday', '17:00', '23:00')
````

````python
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤–µ—á–µ—Ä–∞ –ø—è—Ç–Ω–∏—Ü—ã –≤ –ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ

genre_weekday(spb_general, 'Friday', '17:00', '23:00')
````

2. –í –ú–æ—Å–∫–≤–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ–∫–∞–∑–∞–ª–æ—Å—å —Ç–∞–∫ –º–Ω–æ–≥–æ, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ `'unknown'` –∑–∞–Ω—è–ª–æ –¥–µ—Å—è—Ç–æ–µ –º–µ—Å—Ç–æ —Å—Ä–µ–¥–∏ —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤. –ó–Ω–∞—á–∏—Ç, –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–Ω–∏–º–∞—é—Ç —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—É—é –¥–æ–ª—é –≤ –¥–∞–Ω–Ω—ã—Ö –∏ —É–≥—Ä–æ–∂–∞—é—Ç –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω–æ—Å—Ç–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è.

–í–µ—á–µ—Ä –ø—è—Ç–Ω–∏—Ü—ã –Ω–µ –º–µ–Ω—è–µ—Ç —ç—Ç—É –∫–∞—Ä—Ç–∏–Ω—É. –ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∂–∞–Ω—Ä—ã –ø–æ–¥–Ω–∏–º–∞—é—Ç—Å—è –Ω–µ–º–Ω–æ–≥–æ –≤—ã—à–µ, –¥—Ä—É–≥–∏–µ —Å–ø—É—Å–∫–∞—é—Ç—Å—è, –Ω–æ –≤ —Ü–µ–ª–æ–º —Ç–æ–ø-10 –æ—Å—Ç–∞—ë—Ç—Å—è —Ç–µ–º –∂–µ —Å–∞–º—ã–º.
